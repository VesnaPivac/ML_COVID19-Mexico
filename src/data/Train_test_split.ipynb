# Crear conjuntos de datos de entrenamiento y validación

```python
# Importar librerías
import pandas as pd
import os
from sklearn.model_selection import train_test_split
```

```python
# Cargar los datos procesados
df = pd.read_parquet('../../data/processed/COVID19MEXICO_Processed.parquet')

# Inspeccionar las primeras filas de los datos procesados
df.head()
```

```python
# Separar las variables dependiente e independiente
X = df.drop(columns=['DEFUNCION'])
y = df['DEFUNCION']

# Separar el conjunto de prueba (20% de los datos)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

### Almacenar los datos de entrenamiento y validación

```python
# Ruta de la carpeta donde se guardaran los datos procesadoos
folder_path = '../../data/processed/'

# Verificar si la carpeta no existe y crearla si es necesario
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Guardar datos de entrenamiento
X_train.to_parquet(folder_path + 'X_train.parquet', index=False)
pd.DataFrame(y_train).to_parquet(folder_path + 'y_train.parquet', index=False)

# Guardar datos de validación
X_test.to_parquet(folder_path + 'X_test.parquet', index=False)
pd.DataFrame(y_test).to_parquet(folder_path + 'y_test.parquet', index=False)
```
