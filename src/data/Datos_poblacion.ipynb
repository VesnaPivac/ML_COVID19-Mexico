# Procesamiento de datos poblacionales

```python
import pandas as pd
```

## Cargar datos crudos

```python
# Path al archivo
file_path ='Datos_poblacion.parquet'
# Cargar archivo CSV como DataFrame
df = pd.read_parquet(file_path)
# Inspeccionar primeros elementos del DataFrame
df.head()
```

#### Nos quedamos solo con las columnas de interés

```python
# Analizamos el nombre de las columnas
df.columns 
```

```python
# Crear nuevo Dataframe vacío
df_filtered = pd.DataFrame()

# Extraer columnas de interés del Dataframe original
df_filtered[['CLAVE_ENTIDAD', 'ENTIDAD', 'CLAVE_MUNICIPIO', 'MUNICIPIO', 'LOC', 'NOM_LOC', 'POBLACION']] = df[['ï»¿ENTIDAD', 'NOM_ENT', 'MUN', 'NOM_MUN', 'LOC', 'NOM_LOC', 'POBTOT']]

# Inspeccionamos los datos
df_filtered.head()
```

#### Filtramos para obtener población total por municipio

```python
# Filtrar las filas donde LOC es 0 y NOM_LOC es "Total del Municipio"
df_filtered = df_filtered[(df_filtered['LOC'] == 0) & (df_filtered['NOM_LOC'] == 'Total del Municipio')]

# Seleccionar las columnas de interés
df_filtered = df_filtered[['CLAVE_ENTIDAD', 'ENTIDAD', 'CLAVE_MUNICIPIO', 'MUNICIPIO', 'POBLACION']]

# Inspeccionamos el nuevo Datraframe
df_filtered.head()
```

## Almacenar datos procesados

```python
# Guardar el DataFrame en formato Parquet
df_filtered.to_parquet('Datos_poblacion_processed.parquet', index=False)

print("Archivo guardado como 'Datos_poblacion_processed.parquet'")
```
